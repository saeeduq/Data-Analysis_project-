```python
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(data=df, x='day')
plt.show()
```

```python
sns.countplot(data=df, x='day', order=['Sat', 'Sun', 'Thur', 'Fri'])
plt.show()
```

```python
sns.countplot(data=df, x='day', order=df['day'].value_counts().index)
plt.show()
```

```python
sns.countplot(data=df, x='day', order=df['day'].value_counts(ascending=True).index)
plt.show()
```

```python
sns.countplot(x ='sex', hue = "smoker", data = df)
plt.show()
```

```python
ax = sns.countplot(x ='sex', hue = "smoker", data = df)
# Ø§Ø¸Ù‡Ø§Ø± Ø§Ù„Ø§Ø±Ù‚Ø§Ù… 
for label in ax.containers:
    ax.bar_label(label)
    
plt.show()
```

```python
ax=sns.countplot(x ='sex', palette = "Set2", data = df)

for label in ax.containers:
    ax.bar_label(label)
plt.show()
```

```python
df = pd.read_csv('2016.csv')
df
```

```python
# Ø±Ø³Ù… Ù…Ø®Ø·Ø· Ø§Ù„ØªØ´ØªØª
sns.scatterplot(data = df, x = "Economy (GDP per Capita)", y = "Happiness Score")
plt.show()
```

```python
sns.scatterplot(data = df, x = "Economy (GDP per Capita)", y = "Happiness Score", hue = "Region")
plt.show()
```

```python
plt.figure(figsize=(15,8))

sns.scatterplot(data = df, x = "Economy (GDP per Capita)", y = "Happiness Score", hue = "Region")
plt.show()
```

```python
g = sns.FacetGrid(df, col = "Region", hue = "Region", col_wrap=5)
g.map(sns.scatterplot, "Economy (GDP per Capita)", "Health (Life Expectancy)")
g.add_legend()

plt.show()
```
