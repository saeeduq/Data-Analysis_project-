```python
df = pd.read_csv("StdMarks.csv")
dfcorrel=df[["Math","Phy","Lang"]]
c = dfcorrel.corr(method='pearson')
c
```

رسم مخطط الترابط بين الاعمده 

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
cax = ax.matshow(c, vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = [0,1,2]
ax.set_xticks(ticks)
ax.set_yticks(ticks)
  
ax.set_xticklabels(["Math","Phy","Lang"])
ax.set_yticklabels(["Math","Phy","Lang"])
plt.show()
```
